L’analyse s’appuie sur plusieurs approches complémentaires : la cartographie des localisations Airbnb, l’étude de leur concentration par IRIS, l’analyse ciblée autour d’un pôle attractif comme le MK2 Bibliothèque, ainsi qu’une représentation du prix médian par personne à l’aide d’un maillage régulier de 200 mètres. Ainsi, cette étude s’inscrit dans une logique géomarketing, puisqu’elle vise à mettre en relation la localisation des biens, l’intensité de l’offre, la proximité de lieux attractifs et les niveaux de prix pratiqués. Elle doit permettre à l’entreprise d’orienter sa stratégie d’investissement en repérant les secteurs où l’implantation de biens destinés à Airbnb semble la plus pertinente.

**Démarche méthodologique**

Afin d’identifier les secteurs les plus attractifs pour un investissement locatif de courte durée dans les 12e et 13e arrondissements de Paris, l’étude repose sur une analyse spatiale de données Airbnb réalisée sous Python. La méthodologie adoptée combine préparation des données, traitements géographiques, calculs d’indicateurs et production cartographique.

1. Constitution et préparation des données

L’analyse s’appuie sur un GeoPackage unique contenant plusieurs couches géographiques nécessaires à l’étude : les arrondissements, les IRIS, les parcs, les voies de circulation, les voies ferrées et les points correspondant aux locations Airbnb. Ces données proviennent de sources complémentaires, notamment InsideAirbnb, la BD CARTO® de l’IGN et OpenStreetMap.

Dans un premier temps, l’ensemble des couches est importé dans l’environnement Python à l’aide de bibliothèques spécialisées telles que GeoPandas, Pandas, NumPy, Matplotlib et Shapely. Afin de garantir la cohérence des traitements spatiaux et la fiabilité des calculs de distance et de surface, toutes les couches sont reprojetées dans le système Lambert-93 (EPSG:2154), adapté aux analyses métriques en France.

Cette phase de préparation permet d’obtenir une base de travail homogène, compatible avec les opérations spatiales ultérieures, notamment les jointures géographiques, les calculs de voisinage et la construction d’une grille régulière.

2. Mise en place du fond cartographique et des fonctions de représentation

Avant de réaliser les analyses thématiques, un ensemble de fonctions utilitaires est défini afin de standardiser la production des cartes. Ces fonctions permettent :

d’appliquer un habillage homogène aux cartes ;
de définir automatiquement l’emprise d’affichage avec une marge autour du territoire étudié ;
d’ajouter les étiquettes des arrondissements ;
d’intégrer une barre d’échelle ;
d’ajouter une flèche du nord ;
et d’afficher les crédits cartographiques.

Le fond cartographique est construit à partir des couches de contexte : parcs, réseau ferré, routes, limites d’IRIS et limites d’arrondissements. L’ordre d’affichage des couches est contrôlé par des niveaux de superposition afin d’assurer une bonne lisibilité des cartes produites.

3. Cartographie exploratoire des locations Airbnb

La première étape analytique consiste à représenter les localisations ponctuelles des annonces Airbnb dans les 12e et 13e arrondissements. Cette cartographie exploratoire a pour objectif de fournir une première lecture visuelle de la distribution spatiale de l’offre.

Les points Airbnb sont affichés sur le fond cartographique des deux arrondissements, avec un cadrage automatique sur l’emprise du territoire. Cette première représentation permet de mettre en évidence la concentration générale des annonces, leur dispersion ou leur regroupement dans certains secteurs, ainsi que leur relation avec les principales composantes du tissu urbain.

4. Agrégation des locations par IRIS

Dans un second temps, l’analyse est approfondie à l’échelle des IRIS, qui constituent une maille statistique fine particulièrement adaptée à l’observation des contrastes intra-urbains.

Pour cela, une jointure spatiale est réalisée entre les points Airbnb et les polygones d’IRIS selon une relation d’inclusion spatiale. Chaque annonce est ainsi rattachée à l’IRIS dans lequel elle se situe. À partir de cette correspondance, un comptage est effectué afin de calculer le nombre de locations par IRIS.

Les résultats sont ensuite représentés sous forme de cercles proportionnels. Pour éviter que certains symboles soient placés en dehors des polygones ou dans des zones peu lisibles, les positions de ces cercles sont basées sur des points représentatifs calculés à l’intérieur de chaque IRIS. La taille des cercles est ensuite proportionnée au nombre de locations observé dans chaque unité spatiale.

Cette étape permet d’identifier les quartiers où l’offre Airbnb est la plus dense et de comparer l’intensité de présence des locations à l’échelle infra-arrondissement.

5. Analyse localisée autour du MK2 Bibliothèque

Dans une logique plus directement géomarketing, l’étude intègre ensuite une analyse de voisinage autour d’un pôle d’attractivité urbain, en l’occurrence le MK2 Bibliothèque (François Miterrand), situé dans le 13e arrondissement.

L’adresse du site est d’abord géocodée à l’aide du service OSM **Nominatim** afin d’obtenir ses coordonnées géographiques. Le point correspondant est ensuite reprojeté en Lambert-93, puis un buffer de 600 mètres est construit autour de ce lieu. Cette distance permet de définir un voisinage piéton raisonnable autour du pôle étudié.

Parallèlement, un indicateur économique est calculé pour chaque annonce Airbnb : le prix par personne, obtenu en divisant le prix de la location par la capacité d’accueil (accommodates). Les valeurs aberrantes ou non calculables sont neutralisées.

Les locations incluses dans le buffer sont ensuite sélectionnées, puis le prix médian par personne est calculé pour cet ensemble. Cette étape vise à mesurer le niveau de valorisation de l’offre dans un secteur précis, supposé bénéficier d’une attractivité liée à la présence d’équipements, de transports et d’activités urbaines.

6. Construction d’un maillage régulier de 200 mètres

Afin d’affiner encore l’analyse, une approche par grille régulière est mise en œuvre. Une maille carrée de 200 mètres de côté est générée sur l’ensemble de l’emprise des 12e et 13e arrondissements. Ce choix permet d’obtenir une lecture plus fine des variations spatiales que celle offerte par les limites administratives ou statistiques.

Dans un premier temps, une grille complète est créée à partir des bornes spatiales du territoire. Ensuite, seuls les carreaux intersectant effectivement les arrondissements étudiés sont conservés. Chaque carreau reçoit un identifiant unique afin de faciliter les traitements statistiques.

Cette étape permet de s’affranchir des limites préexistantes et d’analyser les phénomènes spatiaux selon une maille homogène sur tout le territoire.

7. Calcul des indicateurs par carreau

Une nouvelle jointure spatiale est ensuite réalisée entre les points Airbnb et la grille régulière. Cette opération permet de rattacher chaque annonce au carreau dans lequel elle se situe. Deux indicateurs sont alors calculés pour chaque cellule :

le nombre de locations ;
le prix médian par personne.

Les résultats sont ensuite intégrés à la grille. Dans une logique de robustesse statistique, seuls les carreaux comportant plus de cinq locations sont conservés pour l’analyse thématique finale. Ce seuil permet d’éviter de cartographier des valeurs peu représentatives calculées à partir d’un nombre trop faible d’observations.

8. Discrétisation des prix et cartographie finale

Pour rendre la carte interprétable, les valeurs de prix médians par personne sont classées en sept classes selon une méthode fondée sur la moyenne et l’écart-type. Cette discrétisation permet de distinguer les carreaux nettement en dessous de la moyenne, proches de la moyenne ou nettement au-dessus.

Une palette de couleurs progressive est ensuite appliquée afin de produire une carte thématique lisible. Les carreaux sont représentés sur le fond cartographique, avec rappel des limites d’arrondissements, ajout des étiquettes, de l’échelle, de l’orientation et de la légende.

Cette dernière étape permet de visualiser finement les contrastes spatiaux des prix et d’identifier les secteurs les plus valorisés au sein des 12e et 13e arrondissements.

Synthèse de la méthode

Au final, la démarche méthodologique repose sur quatre grands niveaux d’analyse complémentaires :

une lecture exploratoire de la localisation des annonces Airbnb ;
une mesure de la concentration de l’offre par IRIS ;
une analyse localisée autour d’un pôle d’attractivité, le MK2 Bibliothèque ;
une analyse fine des prix grâce à un maillage régulier de 200 mètres.

Cette méthode permet ainsi d’associer une lecture cartographique, statistique et géomarketing du territoire, dans le but d’identifier les secteurs les plus attractifs et les plus pertinents pour un investissement locatif de courte durée.

## Fonctions utilitaires

Les fonctions ci-dessous servent à :
- dessiner le fond de carte sombre ;
- ajouter les labels des arrondissements ;
- dessiner une flèche du nord et une barre d'échelle ;
- enregistrer les figures en PNG.

````python
# -------------------------------------------------------------------
# Fonctions utilitaires de cartographie
# -------------------------------------------------------------------

def add_north_arrow(ax, x=0.95, y=0.90, size=0.08, color="white", zorder=15):
    ax.annotate(
        "N",
        xy=(x, y),
        xytext=(x, y - size),
        arrowprops=dict(arrowstyle="-|>", color=color, lw=1.8),
        ha="center",
        va="center",
        fontsize=12,
        color=color,
        xycoords="axes fraction",
        textcoords="axes fraction",
        zorder=zorder,
    )


def zoom_to_gdf(ax, gdf, margin=0.05):
    """Zoome sur l'emprise d'un GeoDataFrame avec une marge relative."""
    xmin, ymin, xmax, ymax = gdf.total_bounds
    mx = (xmax - xmin) * margin
    my = (ymax - ymin) * margin
    ax.set_xlim(xmin - mx, xmax + mx)
    ax.set_ylim(ymin - my, ymax + my)


def add_scale_bar(
    ax,
    length_m=1000,
    label=None,
    position="bottom_right",
    pad_x=0.08,
    pad_y=0.05,
    color="white",
    zorder=15,
):
    """
    Ajoute une barre d'échelle dans le repère courant.

    position : "bottom_right" ou "bottom_left"
    pad_x / pad_y : marges relatives par rapport au canevas.
    """
    x0, x1 = ax.get_xlim()
    y0, y1 = ax.get_ylim()
    dx = x1 - x0
    dy = y1 - y0

    if position == "bottom_right":
        x_start = x1 - pad_x * dx - length_m
    elif position == "bottom_left":
        x_start = x0 + pad_x * dx
    else:
        raise ValueError("position doit valoir 'bottom_right' ou 'bottom_left'.")

    y_scale = y0 + pad_y * dy
    tick_height = 0.01 * dy
    text_offset = 0.015 * dy

    ax.plot(
        [x_start, x_start + length_m],
        [y_scale, y_scale],
        color=color,
        linewidth=3,
        solid_capstyle="butt",
        zorder=zorder,
    )
    ax.plot(
        [x_start, x_start],
        [y_scale - tick_height, y_scale + tick_height],
        color=color,
        linewidth=2,
        zorder=zorder,
    )
    ax.plot(
        [x_start + length_m, x_start + length_m],
        [y_scale - tick_height, y_scale + tick_height],
        color=color,
        linewidth=2,
        zorder=zorder,
    )

    if label is None:
        label = f"{int(length_m / 1000)} km" if length_m >= 1000 else f"{length_m} m"

    ax.text(
        x_start + length_m / 2,
        y_scale + text_offset,
        label,
        color=color,
        fontsize=10,
        ha="center",
        va="bottom",
        zorder=zorder,
    )


def detect_label_column(gdf, candidates=None):
    """
    Détecte automatiquement une colonne adaptée pour les étiquettes.
    """
    if candidates is None:
        candidates = [
            "l_aroff", "l_ar", "nom", "NOM", "name", "nom_arr", "libelle",
            "libellé", "nomcom", "c_ar", "code", "code_ar", "codearond"
        ]

    for col in candidates:
        if col in gdf.columns:
            return col

    # Repli : première colonne texte hors géométrie avec au moins une valeur non nulle
    for col in gdf.columns:
        if col == gdf.geometry.name:
            continue
        series = gdf[col]
        if series.dtype == "object" and series.notna().any():
            return col

    return None


def add_arrdt_labels(ax, gdf, column=None, color="white", fontsize=10, zorder=20):
    """
    Ajoute les étiquettes des arrondissements au-dessus de toutes les couches.
    """
    labels = gdf.copy()
    labels["geometry"] = labels.representative_point()

    if column is None:
        column = detect_label_column(labels)

    if column is None:
        raise ValueError(
            "Impossible de détecter une colonne d'étiquette pour 'arrdts'. "
            f"Colonnes disponibles : {labels.columns.tolist()}"
        )

    for _, row in labels.iterrows():
        ax.text(
            row.geometry.x,
            row.geometry.y,
            str(row[column]),
            color=color,
            fontsize=fontsize,
            ha="center",
            va="center",
            path_effects=[pe.withStroke(linewidth=2.5, foreground="black")],
            zorder=zorder,
        )


def apply_dark_style(ax):
    ax.set_facecolor("black")
    ax.figure.set_facecolor("black")
    ax.set_axis_off()


def plot_base_map(ax, show_iris=True):
    arrdts.plot(ax=ax, color="none", edgecolor="none", zorder=1)
    parc.plot(ax=ax, color="#3c603c", edgecolor="none", linewidth=0.2, zorder=2)
    rail.plot(ax=ax, color="#545c5c", linewidth=0.2, zorder=3)
    route.plot(ax=ax, color="#272828", linewidth=0.25, zorder=4)

    if show_iris:
        iris.plot(ax=ax, color="none", edgecolor="#7e8585", linewidth=0.5, zorder=5)

    arrdts.plot(ax=ax, color="none", edgecolor="white", linewidth=1.0, zorder=6)
    apply_dark_style(ax)


def add_credits(ax, text):
    ax.text(
        0.01,
        0.01,
        text,
        transform=ax.transAxes,
        color="white",
        fontsize=8,
        ha="left",
        va="bottom",
        zorder=30,
    )
````

## 1. Import des couches et carte des locations Airbnb

````python
fig, ax = plt.subplots(figsize=(12, 8))  # paysage

plot_base_map(ax, show_iris=True)
zoom_to_gdf(ax, arrdts, margin=0.20)

airbnb_l93.plot(
    ax=ax,
    color="darkred",
    markersize=8,
    alpha=0.85,
    zorder=10,
)

ax.set_title(
    "Les locations Airbnb à Paris XII / XIII (2024)",
    color="white",
    fontsize=16,
    pad=12,
)

add_arrdt_labels(ax, arrdts)
add_scale_bar(ax, length_m=2000, label="2 km", position="bottom_right")
add_north_arrow(ax)
add_credits(
    ax,
    "Auteur : Djibril Soumare, 2026\n"
    "BD CARTO®, IGN, 2024 / © Les contributeurs d’OpenStreetMap, 2024 / InsideAirbnb, 2024"
)

output_map_1 = OUTPUT_DIR / "carte_location_airbnb_paris_12_13.png"
fig.savefig(output_map_1, dpi=200, bbox_inches="tight", facecolor=fig.get_facecolor())
plt.show()

print(f"Carte enregistrée : {output_map_1}")
````

## 2. Nombre de locations par IRIS et cercles proportionnels

````python
# -------------------------------------------------------------------
# Nombre de locations par IRIS
# -------------------------------------------------------------------

airbnb_points = airbnb_l93[["geometry"]].copy()

join_iris = gpd.sjoin(
    airbnb_points,
    iris[["geometry"]],
    how="left",
    predicate="within",
)

counts_by_iris = (
    join_iris.groupby("index_right")
    .size()
    .reindex(iris.index, fill_value=0)
    .astype(int)
)

iris = iris.copy()
iris["nb_location"] = counts_by_iris.values

print(iris["nb_location"].describe())
````

````python
def add_north_arrow(ax, position="top_left", zorder=15, fontsize=14):
    x0, x1 = ax.get_xlim()
    y0, y1 = ax.get_ylim()
    dx = x1 - x0
    dy = y1 - y0

    positions = {
        "top_left": (x0 + 0.06 * dx, y1 - 0.12 * dy),
        "top_right": (x1 - 0.06 * dx, y1 - 0.12 * dy),
        "bottom_left": (x0 + 0.06 * dx, y0 + 0.12 * dy),
        "bottom_right": (x1 - 0.06 * dx, y0 + 0.12 * dy),
    }

    x, y = positions.get(position, positions["top_left"])

    ax.annotate(
        "N",
        xy=(x, y),
        xytext=(x, y - 0.08 * dy),
        color="white",
        fontsize=fontsize,
        fontweight="bold",
        ha="center",
        va="center",
        arrowprops=dict(arrowstyle="-|>", color="white", lw=2),
        zorder=zorder,
    )


# Centroides représentatifs pour positionner les cercles
iris_prop = iris.loc[iris["nb_location"] > 0].copy()
iris_prop["geometry"] = iris_prop.representative_point()

# Échelle des symboles
size_factor = 2.8
sizes = iris_prop["nb_location"] * size_factor

fig, ax = plt.subplots(figsize=(12, 8))  # paysage
plot_base_map(ax, show_iris=True)
zoom_to_gdf(ax, arrdts, margin=0.20)

# Cercles proportionnels
ax.scatter(
    iris_prop.geometry.x,
    iris_prop.geometry.y,
    s=sizes,
    color="#a85645",
    edgecolors="white",
    linewidths=0.5,
    alpha=0.85,
    zorder=10,
)

ax.set_title(
    "Distribution des locations Airbnb à Paris XII / XIII (2024)",
    color="white",
    fontsize=16,
    pad=12,
)

# Seules les étiquettes des arrondissements passent au-dessus de tout
add_arrdt_labels(ax, arrdts, zorder=20)
add_scale_bar(ax, length_m=2000, label="2 km", position="bottom_right", zorder=15)
add_north_arrow(ax, position="top_left", zorder=15)

add_credits(
    ax,
    "Auteur : Djibril Soumare, 2026\n"
    "BD CARTO®, IGN, 2024 / © Les contributeurs d’OpenStreetMap, 2024 / InsideAirbnb, 2024"
)

# Légende proportionnelle
legend_values = [160, 50, 2]
legend_handles = [
    plt.scatter(
        [], [],
        s=value * size_factor,
        color="#a85645",
        edgecolors="white",
        linewidths=0.5,
    )
    for value in legend_values
]

legend = ax.legend(
    legend_handles,
    [str(v) for v in legend_values],
    title="Nombre de\nlocations",
    loc="upper right",
    frameon=True,
    facecolor="black",
    edgecolor="white",
    labelcolor="white",
)

plt.setp(legend.get_title(), color="white")

output_map_2 = OUTPUT_DIR / "carte_distribution_airbnb_paris_12_13.png"
fig.savefig(output_map_2, dpi=200, bbox_inches="tight", facecolor=fig.get_facecolor())
plt.show()

print(f"Carte enregistrée : {output_map_2}")
````

## 3. Prix médian par personne autour du MK2 Bibliothèque

Le script R géocode l'adresse `162 Avenue de France, 75013 Paris`, crée un buffer de 600 m, sélectionne les locations incluses dans ce voisinage, puis calcule le prix médian par personne à partir de `price / accommodates`.

Ici, on utilise l'API Nominatim d'OpenStreetMap via `requests` pour effectuer le géocodage sans clé API. Un `User-Agent` explicite est fourni, ce qui est recommandé pour ce service.


````python
# -------------------------------------------------------------------
# Prix médian par personne dans un rayon de 600 m autour du MK2
# -------------------------------------------------------------------

def geocode_nominatim(address, user_agent="notebook-airbnb-paris/1.0"):
    """Géocode une adresse avec l'API Nominatim et renvoie (lon, lat, résultat_brut)."""
    url = "https://nominatim.openstreetmap.org/search"
    params = {
        "q": address,
        "format": "jsonv2",
        "limit": 1,
    }
    headers = {
        "User-Agent": user_agent,
        "Accept-Language": "fr",
    }

    response = requests.get(url, params=params, headers=headers, timeout=30)
    response.raise_for_status()

    results = response.json()
    if not results:
        raise ValueError(f"Aucun résultat trouvé pour l'adresse : {address}")

    result = results[0]
    lon = float(result["lon"])
    lat = float(result["lat"])
    return lon, lat, result


address_mk2 = "162 Avenue de France, 75013 Paris"
lon, lat, geocode_result = geocode_nominatim(address_mk2)

print("Adresse recherchée :", address_mk2)
print("Longitude / Latitude :", lon, lat)
print("Libellé retourné :", geocode_result.get("display_name"))

mk2_wgs84 = gpd.GeoDataFrame(
    {"lieu": ["MK2 Bibliothèque"], "adresse": [address_mk2]},
    geometry=[Point(lon, lat)],
    crs=CRS_WGS84,
)

mk2_l93 = mk2_wgs84.to_crs(CRS_METRIC)

# Calcul du prix par personne
airbnb_l93 = airbnb_l93.copy()
airbnb_l93["prix_pp"] = airbnb_l93["price"] / airbnb_l93["accommodates"]
airbnb_l93.loc[~np.isfinite(airbnb_l93["prix_pp"]), "prix_pp"] = np.nan

# Buffer de 600 m
buffer_600 = mk2_l93.buffer(600)
buffer_600_gdf = gpd.GeoDataFrame(geometry=buffer_600, crs=CRS_METRIC)

# Sélection des locations dans le buffer
within_buffer = airbnb_l93.within(buffer_600_gdf.geometry.iloc[0])
airbnb_600 = airbnb_l93.loc[within_buffer].copy()

prix_median_pp = airbnb_600["prix_pp"].median(skipna=True)

print("Nombre de locations dans le buffer de 600 m :", len(airbnb_600))
print(
    "Le prix médian par personne dans un voisinage de 600 m autour du MK2 Bibliothèque est de",
    round(prix_median_pp, 0),
    "euros."
)
````

## 4. Maillage régulier de 200 m et prix médian par carreau

````python
# -------------------------------------------------------------------
# Création d'une grille régulière de 200 m
# -------------------------------------------------------------------

def make_grid_from_bounds(bounds, cell_size):
    minx, miny, maxx, maxy = bounds
    xs = np.arange(minx, maxx + cell_size, cell_size)
    ys = np.arange(miny, maxy + cell_size, cell_size)

    cells = []
    for x in xs[:-1]:
        for y in ys[:-1]:
            cells.append(box(x, y, x + cell_size, y + cell_size))

    return gpd.GeoDataFrame({"geometry": cells}, crs=CRS_METRIC)


grid = make_grid_from_bounds(arrdts.total_bounds, cell_size=200)
grid["cell_id"] = np.arange(1, len(grid) + 1)

# On conserve uniquement les carreaux qui intersectent les arrondissements
arrdts_union = arrdts.union_all()
grid = grid.loc[grid.intersects(arrdts_union)].copy()

print("Nombre de carreaux après découpe :", len(grid))
````

````python
# -------------------------------------------------------------------
# Nombre de locations et prix médian par carreau
# -------------------------------------------------------------------

points_for_grid = airbnb_l93[["geometry", "prix_pp"]].copy()
points_for_grid["one"] = 1

join_grid = gpd.sjoin(
    points_for_grid,
    grid[["cell_id", "geometry"]],
    how="left",
    predicate="within",
)

stats_grid = (
    join_grid.groupby("cell_id")
    .agg(
        nb_location=("one", "sum"),
        prix_median_pp=("prix_pp", "median"),
    )
    .reset_index()
)

grid = grid.merge(stats_grid, on="cell_id", how="left")
grid["nb_location"] = grid["nb_location"].fillna(0).astype(int)
grid["prix_median_pp"] = grid["prix_median_pp"].astype(float)

grid_5 = grid.loc[grid["nb_location"] > 5].copy()

print(grid_5[["nb_location", "prix_median_pp"]].describe())
````

Pour l'analyse de transactions, on garde uniquement les carreaux comportant plus de 5 transactions, puis applique une discrétisation en 7 classes basée sur l'écart-type pour cartographier `prix_median_pp`. On retient 7 classes construites autour de la moyenne et de l'écart-type.

````python
# -------------------------------------------------------------------
# Discrétisation en 7 classes autour de la moyenne et de l'écart-type
# -------------------------------------------------------------------

def std_mean_breaks(series):
    s = pd.Series(series).dropna()
    mu = s.mean()
    sigma = s.std(ddof=1)

    # 7 classes : <=-2σ ; -2σ/-1σ ; -1σ/-0.5σ ; -0.5σ/+0.5σ ;
    # +0.5σ/+1σ ; +1σ/+2σ ; >+2σ
    return [-np.inf, mu - 2*sigma, mu - sigma, mu - 0.5*sigma,
            mu + 0.5*sigma, mu + sigma, mu + 2*sigma, np.inf]


def format_interval_labels(bins):
    labels = []
    for left, right in zip(bins[:-1], bins[1:]):
        left_txt = "-∞" if np.isneginf(left) else f"{left:.1f}"
        right_txt = "+∞" if np.isposinf(right) else f"{right:.1f}"
        labels.append(f"[{left_txt} ; {right_txt}]")
    return labels


bins = std_mean_breaks(grid_5["prix_median_pp"])
labels = format_interval_labels(bins)

grid_5["classe_prix"] = pd.cut(
    grid_5["prix_median_pp"],
    bins=bins,
    labels=labels,
    include_lowest=True,
)

palette = [
    "#d9f0d3",
    "#addd8e",
    "#78c679",
    "#ffffbf",
    "#fed976",
    "#fd8d3c",
    "#e6550d",
]

color_map = dict(zip(labels, palette))
grid_5["couleur"] = grid_5["classe_prix"].map(color_map)
````
